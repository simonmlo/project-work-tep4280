## $$\text{Visualization of Drag Coefficient}$$

CylinderCase1Data.dat = First run (199s) on icoFoam

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("Data/baseTest.dat", sep=r"\s+")
df2 = pd.read_csv("Data/baseTestReduction.dat", sep=r"\s+")

time = df["Time"]
Cd = df["Cd"]
Cl = df["Cl"]

time2 = df2["Time"]
Cd2 = df2["Cd"]
Cl2 = df2["Cl"]

# Create figure
fig, ax = plt.subplots(1, 2, figsize=(14, 5), constrained_layout=False)
fig.suptitle("Introducing an initial velocity field", fontsize=16)

# --- Drag plot ---
ax[0].plot(time, Cd, label="Cd")

# Final point
Cd_last = Cd.iloc[-1]
t_last = time.iloc[-1]

Cd_last2 = Cd2.iloc[-1]
t_last2 = time2.iloc[-1]

ax[0].scatter(t_last, Cd_last, color='r', marker='x', label=f"Final: {Cd_last:.3f}")


ax[0].set_title("Full 200s simulation (CPU-time 289s)")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Cd")
ax[0].grid(True)
ax[0].legend()

# --- Lift plot ---
ax[1].plot(time2, Cd2, label="Cd")

# Cl_last = Cl.iloc[-1]

ax[1].scatter(t_last2, Cd_last2, color='r', marker='x', label=f"Final: {Cd_last2:.3f}")

ax[1].set_title("Reduction to 50s simulation (CPU-time 73s)")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Cl")
ax[1].grid(True)
ax[1].legend()

plt.show()
```
