## $$\text{Comparison of Drag Coeffs. between two simulations}$$

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df1 = pd.read_csv("Data1", sep=r"\s+")
df2 = pd.read_csv("Data2", sep=r"\s+")

time1 = df1["Time"]
Cd1 = df1["Cd"]

time2 = df2["Time"]
Cd2 = df2["Cd"]

# Create figure
fig, ax = plt.subplots(1, 2, figsize=(14, 5), constrained_layout=False)
fig.suptitle("Comparison Sim 1 v.s. Sim 2", fontsize=16)

# --- Drag plot ---
ax[0].plot(time1, Cd1, label="Cd")

# Final point
Cd_last = Cd1.iloc[-1]
t_last = time1.iloc[-1]

Cd_last2 = Cd2.iloc[-1]
t_last2 = time2.iloc[-1]

ax[0].scatter(t_last, Cd_last, color='r', marker='x', label=f"Final: {Cd_last:.3f}")


ax[0].set_title("<Title Sim 1>")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Cd")
ax[0].grid(True)
ax[0].legend()

# --- Lift plot ---
ax[1].plot(time2, Cd2, label="Cd")

# Cl_last = Cl.iloc[-1]

ax[1].scatter(t_last2, Cd_last2, color='r', marker='x', label=f"Final: {Cd_last2:.3f}")

ax[1].set_title("<Title Sim 2>")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Cd")
ax[1].grid(True)
ax[1].legend()

plt.show()
```
