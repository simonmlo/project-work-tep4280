# $$\text{Steady State control}$$

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df1 = pd.read_csv("controlSteadyState/baseCase.dat", sep=r"\s+")
df2 = pd.read_csv("controlSteadyState/firstTryVelocity.dat", sep=r"\s+")
df3 = pd.read_csv("controlSteadyState/secondTryVelocity.dat", sep=r"\s+")
df4 = pd.read_csv("controlSteadyState/thirdTryVelocity.dat", sep=r"\s+")


t1  = df1["Time"] 
Cd1 = df1["Cd"]

t2  = df2["Time"] 
Cd2 = df2["Cd"]

t3  = df3["Time"] 
Cd3 = df3["Cd"]

t4  = df4["Time"] 
Cd4 = df4["Cd"]

startTime = 10
stopTime  = 2000

timeAverage1 = np.round(np.mean(Cd1[1000:2000]), 5)
timeAverage2 = np.round(np.mean(Cd2), 5)
timeAverage3 = np.round(np.mean(Cd3), 5)
timeAverage4 = np.round(np.mean(Cd4), 5)

plt.figure(figsize=(18, 6))


plt.plot(t1[startTime:stopTime], Cd1[startTime:stopTime], label=f'13 248 cells {timeAverage1}')
plt.plot(t2, Cd2, label=f'13 248 cells {timeAverage2}')
# plt.plot(t3[2:], Cd3[2:], label=f'13 248 cells {timeAverage3}')
plt.plot(t4, Cd4, label=f'13 248 cells {timeAverage4}')

plt.tight_layout()
plt.legend()
plt.grid()
plt.show()
```
