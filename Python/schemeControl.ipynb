# $$\text{Gauss Linear vs. Gauss Minmod (a TVD-scheme)}$$

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Loading data
GaussLinear     = pd.read_csv("schemeData/x3Linear.dat", sep=r"\s+")
GaussMinmod     = pd.read_csv("schemeData/x3Minmod.dat", sep=r"\s+")
GaussvanLeer    = pd.read_csv("schemeData/x3vanLeer.dat", sep=r"\s+")
GaussLimitedLinear1 = pd.read_csv("schemeData/x3Limited.dat", sep=r"\s+")
GaussLimitedLinear2 = pd.read_csv("schemeData/x3Limited2.dat", sep=r"\s+")

timeLinear  = GaussLinear["Time"]
timeMinmod  = GaussMinmod["Time"]
timevanLeer = GaussvanLeer["Time"]
timelimitedLinear1 = GaussLimitedLinear1["Time"]
timelimitedLinear2 = GaussLimitedLinear2["Time"]

cdLinear    = GaussLinear["Cd"]
cdMinmod    = GaussMinmod["Cd"]
cdvanLeer   = GaussvanLeer["Cd"]
cdLimitedLinear1 = GaussLimitedLinear1["Cd"]
cdLimitedLinear2 = GaussLimitedLinear2["Cd"]

start = 15
stop = 50000

fig = plt.figure(figsize=(12, 7))

fig.suptitle('Visualization of how different schemes effect the solution')

plt.plot(timeLinear[start:], cdLinear[start:], label=r'Gauss Linear,  CFL $\approx$ 0.12764, CPU-time = 147s')
plt.plot(timeMinmod[start:], cdMinmod[start:], label=r'Gauss Minmod, CFL $\approx$ 0.124177, CPU-time = 156s')
plt.plot(timevanLeer[start:], cdvanLeer[start:], label=r'Gauss vanLeer, CFL $\approx$ 0.127613, CPU-time = 170s')
plt.plot(timelimitedLinear1[start:], cdLimitedLinear1[start:], label=r'Gauss Limited Linear 1, CFL $\approx$ 0.125687, CPU-time = 156s')
plt.plot(timelimitedLinear2[start:], cdLimitedLinear2[start:], label=r'Gauss Limited Linear 0.75, CFL $\approx$ 0.125014, CPU-time = 156s')

plt.tight_layout()
plt.xlabel('t')
plt.ylabel('Cd')
plt.legend()
plt.show()
```
