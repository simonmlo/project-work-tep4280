## $$\text{Drag Coeff. Comparison}$$

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df1 = pd.read_csv("forceCoeff/x3delta10x.dat", sep=r"\s+")
df2 = pd.read_csv("forceCoeff/x4.dat", sep=r"\s+")
df3 = pd.read_csv("forceCoeff/x5.dat", sep=r"\s+")

time1 = df1["Time"]
Cd1 = df1["Cd"]

time2 = df2["Time"]
Cd2 = df2["Cd"]

time3 = df3["Time"]
Cd3 = df3["Cd"]

start = 100

# Create figure
fig, ax = plt.subplots(1, 3, figsize=(20, 6), constrained_layout=False)
fig.suptitle("Comparison Between Different Meshes", fontsize=16)

# --- Plot 1 ---
ax[0].plot(time1[start:], Cd1[start:], label="Cd")

# Final point
Cd_last = Cd1.iloc[-1]
t_last = time1.iloc[-1]

Cd_last2 = Cd2.iloc[-1]
t_last2 = time2.iloc[-1]

Cd_last3 = Cd3.iloc[-1]
t_last3 = time3.iloc[-1]

ax[0].scatter(t_last, Cd_last, color='r', marker='x', label=f"Final: {Cd_last:.3f}")

ax[0].set_title("5748 Cells")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Cd")
ax[0].grid(True)
ax[0].legend()

# --- Plot 2 ---
ax[1].plot(time2[start:], Cd2[start:], label="Cd")

# Cl_last = Cl.iloc[-1]

ax[1].scatter(t_last2, Cd_last2, color='r', marker='x', label=f"Final: {Cd_last2:.3f}")

ax[1].set_title("13248 Cells")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Cd")
ax[1].grid(True)

ax[1].legend()

# --- Plot 3 ---
ax[2].plot(time3[start:], Cd3[start:], label="Cd")

# Cl_last = Cl.iloc[-1]

ax[2].scatter(t_last3, Cd_last3, color='r', marker='x', label=f"Final: {Cd_last3:.3f}")

ax[2].set_title("32500 Cells")
ax[2].set_xlabel("Time")
ax[2].set_ylabel("Cd")
ax[2].grid(True)

ax[2].legend()

plt.show()
```
